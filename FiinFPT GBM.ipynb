```python
#GeoBrownian
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
price = pd.read_csv('FPT.csv')
price['Close'] = price['Close'].astype('int64')
```

```python
# Tnh li sut log
log_returns = np.log(price['Close'] / price['Close'].shift(1))

# Tnh drift (trung bnh li sut)
mu = log_returns.mean()

# Tnh volatility ( lch chun ca li sut)
sigma = log_returns.std()

# Annualize (chuyn i sang n v nm)
mu_annualized = mu * (252 * 6.5 * 60)  # 252 ngy, mi ngy c 6.5 gi, mi gi c 60 pht
sigma_annualized = sigma * np.sqrt(252 * 6.5 * 60)
# Cc thng s
S0 = price['Close'].iloc[-1]  # Gi cui cng trong d liu hin ti
T = 252  # S ngy giao dch cn m phng
n_intervals_per_day = int(6.5 * 60)  # S khong 1 pht trong mt ngy giao dch
n_steps = T * n_intervals_per_day  # Tng s bc thi gian
dt = 1 / n_steps  # Bc thi gian trong n v nm
n_scenarios = 100  # S kch bn m phng
# t random seed  ti lp kt qu
np.random.seed(42)

# Tnh drift v shock
drift = (mu_annualized - 0.5 * sigma_annualized**2) * dt
shock = sigma_annualized * np.sqrt(dt)

# To ma trn kch bn ngu nhin cho cc bc thi gian
Z = np.random.normal(0, 1, (n_steps, n_scenarios))
random_walk = drift + shock * Z

# Tnh gi m phng cho tng kch bn
price_paths = np.zeros((n_steps + 1, n_scenarios))
price_paths[0] = S0  # Gi ban u

for t in range(1, n_steps + 1):
    price_paths[t] = price_paths[t - 1] * np.exp(random_walk[t - 1])

# V  th kt qu m phng
plt.figure(figsize=(12, 6))
for i in range(min(n_scenarios, 1000)):  # Hin th ti a 1000 kch bn
    plt.plot(price_paths[:, i], lw=1)
plt.title('Geometric Brownian Motion Simulation')
plt.xlabel('Time Steps')
plt.ylabel('Simulated Price')
plt.show()
```

```python
# Tnh bin ng gi (Delta P)
delta_price = np.diff(price_paths, axis=0) #m phng gi c phiu tng pht v mun tnh bin ng gi gia hai pht lin tip 
# Tnh t l bin ng gi (Delta P%)
delta_pct = (delta_price / price_paths[:-1, :]) * 100
mean_delta_price = delta_price.mean(axis=1)  # Trung bnh bin ng gi theo tng bc
std_delta_price = delta_price.std(axis=1)    #  lch chun bin ng gi theo tng bc
# Phn phi bin ng gi ti bc thi gian c th (v d, ngy u tin)
plt.hist(delta_price[0], bins=30, alpha=0.7, color='blue')
plt.title('Distribution of Price Changes (Day 1, 1-min intervals)')
plt.xlabel('Price Change')
plt.ylabel('Frequency')
plt.show()
```

```python
# Biu  xu hng trung bnh v  lch chun
plt.figure(figsize=(12, 6))
plt.plot(mean_delta_price, label='Mean Price Change', color='blue')
plt.fill_between(range(len(mean_delta_price)), 
                 mean_delta_price - std_delta_price, 
                 mean_delta_price + std_delta_price, 
                 color='blue', alpha=0.2, label='Standard Deviation')
plt.title('Trend of Price Changes Over Time')
plt.xlabel('Time Steps (1-min intervals)')
plt.ylabel('Price Change')
plt.legend()
plt.show()
```
